```python
# Comparison on games dealt, iOS only

mu_off = 133.61
mu_on = 158.72
mu_off_new = 115.40
mu_on_new = 115.92
mu_off_exist = 199.22
mu_on_exist = 209.09

se_off = 1.04
se_on = 0.96
se_off_new = 0.99
se_on_new = 0.98
se_off_exist = 3.15
se_on_exist = 1.71
```

```python
mu = [mu_off, mu_on, mu_off_new, mu_on_new, mu_off_exist, mu_on_exist]
se = [se_off, se_on, se_off_new, se_on_new, se_off_exist, se_on_exist]
names = ['Off_Overall', 'On_Overall', 'Off_New', 'On_New', 'Off_Exist', 'On_Exist']
```

```python
from __future__ import print_function, division
import numpy as npfrom scipy.stats import norm

plt.figure(figsize=(50, 20))
plt.xlabel('avg games dealt', fontsize = 40)
plt.xticks(fontsize = 40)
plt.yticks(fontsize = 40)
plt.title('Sampling Distribution of Avg Games Dealt by Cohort', fontsize = 40)
for m, s, n in zip(mu, se, names):
    x = np.linspace(m+3*s, m-3*s, 200)
    y = norm.pdf(x, m, s)
    plt.plot(x, y, linewidth = 10)
    plt.vlines(x=m, ymin = 0, ymax = max(y), linestyles='dashed', label = n)
    plt.text(m, max(y)+0.001, n, fontsize = 30)

    plt.tight_layout()
    plt.savefig('castle_avg_game_dealt_dist.png')
```
